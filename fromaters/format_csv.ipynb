```python
import csv
import json
import pandas as pd
import os
```

```python
def excel_to_csv(excel_file):
    try:
        df = pd.read_excel(excel_file)
        csv_file = excel_file.split('.')[0] + '.csv'
        df = df.drop(['Unnamed: 0'],axis=1)
        df.to_csv(csv_file, index=False)
        if os.path.exists(csv_file):
            return csv_file
        else:
            return "CSV file creation failed."

    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
def csv_to_json(csv_file, json_file,url):
    with open(csv_file, 'r') as f:
        reader = csv.DictReader(f)
        data = []
        for row in reader:
            json_row = {}
            content = ', '.join([f"{key} : {value}" for key, value in row.items()])
            raw_title = str(csv_file.split('\\')[-1])
            json_row['title'] = raw_title.split(".")[0]
            json_row['content'] = content
            json_row['url'] = url
            json_row['doc_name'] = raw_title
            
            data.append(json_row)
    
    with open(json_file, 'w') as f:
        json.dump(data, f, indent=4)
```

```python

csv_path = r'E:\Projects\SA - R&D\RAG with CSV\data\file_example_XLSX_1000.xlsx'
op_path = r'E:\Projects\SA - R&D\RAG with CSV\data\output.json'
url = "https://drive.google.com/file/d/1yn_5Nauy2QoHhRcdxlMpWH8zsd2Q5REo/view?usp=sharing"

if csv_path.lower().endswith('.csv'):
  print("Given file is a CSV")
  csv_to_json(csv_path, op_path,url)
  

elif csv_path.lower().endswith('.xls') or csv_path.lower().endswith('.xlsx'):
  print("Given file is not a csv")
  csv_path = excel_to_csv(csv_path)
  csv_to_json(csv_path,op_path,url)

else:
  print("Invalid file expected a CSV or XLS or XLSX")
```
